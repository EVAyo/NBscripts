# 读《数字原住民网络潜水动因实证研究》

## 研究目的及意义

研究目的（微观）：

1. 界定潜水者和数字原住民型潜水者消除对于潜水的模糊定义。
1. 构建潜水动因模型提出各项解释数字原住民潜水行为的动因假设。
1. 明确主要影响因素以及因素之间的相互影响关系。
1. 为网络运营商和社区管理者提供更有效的建议。 

研究意义（宏观）：

一、理论意义：梳理社会学、经济学和心理学三大学科中用于潜水研究的相关理论并搭建了潜水研究的理论框架为后续研究者的相关研究提供了理论支撑。<br>
二、实际意义：对于数字原住民潜水行为研究的实际意义主要体现在经济效益、网络可持续发展以及群体智慧三方面。

## 网络潜水研究述评

### 概念定义及研究进展

#### 数字原住民定义

数字原住民(Digital Natives)的概念最早是由Marc Prensky于2001年提出来的，用于区别于数字移民(Digital Immigrants)。在他的研究文献中，将数字原住民定义为和高科技一起诞生、学习、生活并长大成人者，也就是一出生就面临网络世界的年轻人;而将数字移民定义为年龄较大，在新兴的信息科技、数字技术面前需要通过困难或不通畅的学习才能掌握的群体。(刘江,2013,p9)

#### 网络潜水者定义

据刘江[<span id="fn1">(2013,p12,p13)</span>](#fn1-black)对网络潜水者的细分如下：

**消极沉默的潜水者(Passive Lurkers-P):** 这是一类传统意义上的潜水者，他们彻底遵循只看不说的原则，有时也只阅读对自己有用的信息和内容，从不创作内容或共享内容等。

**积极活跃的潜水者(Active lurker-A):** 这一类潜水者在网络社区中并不创作内容或共享内容或参与讨论或评论,而总是直接与创作者通过邮件或者其他即时聊天工具等进行私下相五交流，谈论自己的看法或获取对自己有用的东西。

**间歇性的潜水者(Occasional lurkers-OL):** 这一类潜水者是以时间轴为基准来定义的，即在排除正常的个人因素或环境因素(如工作忙、不方便上网、电脑损坏等等一系列因素)后，某些网络用户在其行为纵向上时而活跃时而沉寂。间隙性潜水者通常可用于帮助一个网络社区或者社会化媒体的管理人员观察发掘其用户流失、用户黏性和忠诚度等状况。

**浮出水面的潜水者(De-lurkers-DL):** 是指网络社区中潜水观望了一段时期后，由于某种原因诱导开始在社区论坛中创作、共享内容或参与讨论的用户群体。当然这种潜水者并不包括一开始进入创作、共享内容或参与讨论的用户。

### 潜水研究的动因体系

目前国外研究者们对外在动因的研究主要采用**问卷调查、访谈和实验**的方法进行研究，而提取出来的外在动因也比较丰富。[<span id="fn1">(刘江,2013,p18)</span>](#fn1-black)

#### 内在动因

内在动因是从潜水者潜水行为的内部影响因素来探索潜水者的潜水动因，如从个人心理、个性特点、知识技能等的角度来分析潜水者的潜水行为内部驱动的影响因素。内在动因具有极强的内在化和自主性，即人的内在心理特征、情绪和认知结构强烈影响了实际行为模式。通过文献梳理内在动因主要表现在经济利益层面、社会层面以及知识技术层面。内在动因经济利益层面上的折射相对较少，主要体现在个人需求与满足方面。[<span id="fn1">(刘江,2013,p16)</span>](#fn1-black)

一、安全隐私：泄露自己的隐私信息或给自己带来不安全而选择潜水<br>
二、个性特点：害羞并缺乏自信从而不好意思参与发帖 <br>
三、时间限制：个人时间有限从而每次进入社区论坛时只看不说<br>
四、发帖即承诺：社群未达到自我的期望，即归属感<br>
五、知识技术：专业知识匮乏和个人能力欠缺<br>


#### 外在动因

潜水者的外在动因研究是从外部影响因素来探索潜水者的潜水动因,包括人为因素、组织因素、环境(情境)因素和社会因素的影响。外在动因的研究相对而言可操作性强，研究起来也较为方便。而外在动因也主要表现在经济利益层面、社会层面以及知识技术层面。外在动因在经济利益层面的突出表现便是外部奖励。外部奖励是指社区论坛网站等通过对发布者或评论者进行物质、金钱或者其他形式的回报奖励从而诱发网络用户的参与热情来影响网络参与者的相关行为。[<span id="fn1">(刘江,2013,p18)</span>](#fn1-black)

一、信息质量：信息质量较低、资源匮乏不能引起参与者兴趣，或者信息过载让参与者感到厌烦和累赘。<br>
二、工作限制：入群氛围不合时宜，或网站屏蔽，部分信息被审查。<br>
三、组织认同与氛围：不认同参与者、或组织氛围不友好。<br>
四、技术壁垒：早期的社论坛中由于技术发展的限制对于发布者技术要求很高，需要发布者知晓相关的计算机语言或网页语言<br>
五、交互延迟：长时间看不到自己的帖子或基本上没什么互动性。<br>
六、外部奖励：资金奖励，但取决于个人参与度与认同度，否则再大的奖励也只是昙花一现。<br>

#### 激励措施

这里提出我的看法：

一、安全方面：加强宣传与倡导，并一些不良涉及隐私的案例做为论坛对安全方面的重视。<br>
二、个性特点：设置“圈子”、组群的讨论组交流。<br>
三、时间方面：鼓励用户建立综合性话题讨论。<br>
四、发帖方面：设置专业区与开放性的“无所不谈，畅所欲言”的水区<br>
五、学识方面：论坛方面应该出一些教学视频辅导潜水者。让他们有一定的基础知识沉淀。<br>
六、外部奖励：给予善于发帖者及活跃者特殊“勋章”，设置生日信、升级信、以及为他们构建进一步的圈子交流。<br>

刘江的观点：

一、正视成员动机差异，采取差异化激励措施：有必要以年龄为基础对社区或网络用户群体进行划分，对数字原住民群体和数字移民群体有针对性的采取不同的激励措施。<br>
二、增强外部激励和社区互动性，削弱数字原住民用户潜水行为意向，刺激其积极参与：成员等级晋升、积分回馈制度、社区明星榜或活跃榜，对入榜成员进行物质奖励。<br>
三、加强数字原住民的虚拟社区感的培养：对网暴言行进行风控，创建和谐友好的社区文化氛围；定期举办活动评选，加强凝聚力与归属感；加强社区成员间的信任感建立起实名注册。<br>
四、定期邮件调研及时了解社区成员潜水动因：咨询其潜水的相关动因或对于社区网站的不满意之处。<br>

## 理论模型构建与假设提出

### 潜水研究的理论基础

前期主要鉴定潜水员是否是合法用户，如“潜水员只在享受别人的分享”、“局外人”、“属于论坛的合规成员”。随着后续有了社会学、社会心理学、组织行为学、经济学等学科领域做支撑，潜水现象及行为也有了更进一步细分。刘江[<span id="fn1">(2013,p20)</span>](#fn1-black)对此进行了具体的论证：如 Cliff Lampe基于社会心理学中的社会懈怠效应来探究社区论坛用户群体规模的扩增对用户的贡献积极性和搭便车情绪的影响，Rafaeli等则引入社会资本理论来诠释网络潜水行为与社会政治参与行为的相似性，并且还引入该理论定义虚拟社区用户的社会资本和构建社交网络活动，从而探究个人社会资本对于自己或其他用户在虚拟社区中参与程度的影响。

### 模型构建的理论基础

#### 计划行为理论

时间标志：Ajzen在1991年发表的《The Theory of Planned Behavior》一文中对计划行为理论进行了进一步的阐释，同时也标志着该理论逐步走向成熟和应用。

刘江[<span id="fn1">(2013,p27)</span>](#fn1-black)将“计划行为理论”的抽象概念进行逐一阐述并以图示将概念流程分解与关联，他认为计划行为理论模型构成的五要素是行为态度，主观规范，知觉行为控制，行为意向以及实际行为，各个变量的含义如下：
* 态度：指个体对执行相关行为所持的正面或负面的相关感觉，也是指由个体对此特定行为的相关评估经过概念化之后而形成的态度。
* 主观规范：指周围社会环境压力对个体行为抉择的影响，它反映了对介体具有重要性的其他个体和团队对其行为决策的影响。
* 知觉行为控制：指个体依据过去的经验，所预期到的执行某种特定行为的阻碍或难易程度。
* 行为意向：指个体对于执行某项特定行为的主观机率的判定，反映了个体执行某一项特定行为的意愿程度。
* 实际行为：是指个体实际采取行动的行为。

#### 社会资本理论

我的想法，其通俗的解释是：群体成员对社会公共环境的各项期望，由此衍生完整的社会资本环境的构念。（**合作效率，配合度**）

刘江[<span id="fn1">(2013,p27)</span>](#fn1-black)对社会资本论总结的主要观点有：

1. 社会资本能对群体成员产生正向的外部效应；
2. 这些外部效应能够通过共享信任、规范、价值观念以及对预期和行为产生影响的相关结果来获得。
3. 共享信任、规范、价值观念产生于基于社会网络和关系的
4. 相关组织的非正式形式。而对社会资本的研究就是探究通过信任、规范产生有益结果的关系网络的相关流程。

在个体间关系的研究中，社会资本理论的应用范围已得到了进一步的延伸和扩展，以此来探索如何通过团队协调促进社会效率。

### 问卷设计

**批注：符合论文写作的过程 (p38)**

具体执行过程：在研究中数据收集是通过调查问卷来实现的，所以问卷设计质量对所收集到的数据质量有很重要的影响。在问卷设计阶段，首先设计问卷的总体框架，针对本次研究模型中10个潜变量并基于前人研究基础上设计测量问项，进而构建本研究的初始调查问卷。接着通过调研访谈和小规模前测，修改易造成歧义的问项，并利用前测数据进行问卷的信度和效度检验。最后，根据检验结果对测量问项进行进一步的修正，形成正式问卷并发放，收集研究数据。

* **信度分析：主要是考察各潜变量的Cronbach'sa值是否超过0.7的临界值。**
* **效度分析，主要基于因子分析来检验。也就是通过对样本数据进行KMO系数分析。Kaiser给出了常用的kmo度量标准:　0.9以上表示非常适合；0.8表示适合；0.7表示一般；0.6表示不太适合；0.5以下表示极不适合。**
* **因子相关性分析：Bartlett球形检验则用于检验观测变量之间是否存在相关性，这是进行因子分析的一个前提条件。它基于一个假设，即观测变量之间不存在任何相关性。如果在Bartlett球形检验中得出的显著性水平较低（通常设置为0.05），则可以拒绝该假设，表明观测变量之间存在相关性，因此适合进行因子分析。**
* **偏向程度分析：，Likert量表的评分范围为5或7个等级，但也可以是其他数字。选择适当的评分范围要考虑到被调查者容易理解和回答的程度。较少的等级可能更容易导致患者在选择中立选项时受到限制，而较多的等级可能增加了患者选择的复杂性。**

经过问卷前测和信度、效度分析，问卷的各个问项均满足相关标准，所以形成正式问卷。问卷是基于匿名调查的，所受到社会称许性偏差较小。(问卷图见，p69)

**问卷发放样本值选定值**

有关结构方程分析中样本规模标准，国内外研究者们的意见并不统一。如研究者Anderson等认为进行结构方程模型分析的最低样本数目为100-150'；邱皓政提出结构方程分析的样本数目不能低于200以保证稳定的分析结果。而Gorsuch的观点相对而言被大多数研究者所认可并加以采用，他认为样本规模应为测量问项的5到10倍，并且100是样本规模的最低阀值。基于以上研究者的结论，刘江将本研究的最低样本数目定为200，以便于进行稳定且准确的结构方程分析。

本研究从2012年3月21日到2012年3月31日发放和回收问卷，本次问卷发放共回收问卷245份，主要通过邮件发送和即时通讯软件发送两种方式发放问卷。并依照以下标准剔除无效问卷：<br>
(1)答案选项过于集中；(2)问卷填写不完整，有遗漏项；(3)问卷填答有矛盾的答案；<br>
根据以上标准，筛选有效问卷，得到215份有效问卷，即问卷有效率为87.6%。本研究将基于这215份样本数据进行数字原住民网络潜水动因的实证分析研究。

## 数据分析与假设检验

统计方法在数据分析中起到了校对和辅助参考的作用。它们帮助我们对数据进行整理、总结和解释，并提供了一种系统性的方式来推断总体特征和进行决策。然而，统计方法也有其限制和假设，需要谨慎应用。例如，在因子分析中，样本大小对于因子的稳定性和可靠性有一定影响。较小的样本容易导致结果的不稳定性和误导性，因此需要在进行因子分析时考虑样本的大小，并确保样本足够大以支持可靠的结果。

此外，统计分析还依赖于所使用的数据质量和采样方法的正确性。如果样本选择不当、数据存在缺失或偏差等问题，统计分析的结果可能会出现偏差或误导。因此，在进行统计分析之前，需要仔细检查数据的可靠性和有效性，并在可能的情况下采取纠正措施。

总的来说，统计方法是有用的工具，可以提供对数据的分析和解释。然而，在使用统计方法时，我们需要理解其局限性，并谨慎地解释和应用结果，避免过度依赖统计分析的结果而忽略实际情况的综合考量。

## 结语

研究梳理国内外学者对网络潜水行为研究成果的基础上，对网络潜水者更近一步细分、整合。构建潜水行为动因分类体系，整合多门学科，基于计划行为理论、社会交换论、社会资本论以及社会认同理论，挖掘数字原住民动因潜水模型，并应用统计学实证对模型进行检验。

### 研究结论

(1)数字原住民潜水动因与数字移民潜水动因之间存在一定的差异性。<br>
(2)虚拟社区中的互惠性和交五性均对数字原住民的潜水行为意向有显著的负向影响。<br>
(3)虚拟社区感对虚拟社区中数字原住民潜水行为意向和潜水行为均有显著的负向影响。<br>
(4)社区的文化氛围、成员的社区认同感以及社区成员间的信任对数字原住民的虚拟社区感有显著的正向影响。<br>
(5)物理限制条件也是数字原住民潜水行为的显著影响因素。<br>

### 实践建议

(1)正视成员动机差异，采取差异化激励措施<br>
(2)增强外部激励和社区互动性，削弱数字原住民用户潜水行为意向，刺激其积极参与<br>
(3)加强数字原住民的虚拟社区感的培养<br>
(4)定期邮件调研，及时了解社区成员潜水动因<br>

### 研究局限

(1)模型构建方面：研究成果较少，模型理论支撑度较单薄。<br>
(2)问卷设置方面：研究成果较少，问卷内容可能存在缺陷，待改进。（开创性考量）<br>
(2)调查样本方面：样本数偏少，后续需进一步扩大规模，增强说服力。（时限，规模）

### 研究展望

后续研究者可进一步完善的研究：

1. 将动因模型用于数字原住民与数字移民做比较，深入理解潜水因素。
2. 增加调节变量（年龄、职业、教育背景）进行进一步衡量潜水影响因素。
3. 增加文献的查阅，增强样本的代表性与普适性。
