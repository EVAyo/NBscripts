## 民生类新闻，以《1818黄金眼》为例

虽然我本人看《1818黄金眼》的次数较少，目前总的算下来也就看了五回左右不等。以下为了方便叙述，对《1818黄金眼》简称《黄金眼》。就我自身而言，观看时间基本上聚焦于《黄金眼》提供的内容，未有在《黄金眼》选材、选题方面的思索。再回顾《黄金眼》往期内容，《黄金眼》新闻标题通常以疑问句，或事件争端的陈述句为主；视频时长通常在十分以内，一天发布的2～3部及以上视频。[<span id="fn1">覃玉颖（2021.8）</span>](#fn1-black)的观点《黄金眼》在传统的民生新闻上，增加了对街头趣闻、市井奇闻、名人轶事的集中发布，让用户本能地频繁切换信息元素，满足自我猎奇心理，呈现碎片化内容生产。总体上对《黄金眼》的基本信息概括相符。

不同地域的人，在单向度的现代社会中，往往面临着相似的 生活困境与社会纠纷，经由媒体选择典型民生个例加以报道，很容易使民众获得切身体会，产生跨地域的情感共鸣，这意味着地 方民生类新闻只要在选材上能抓住时代共性、传播上能够满足时代的要求，就具备了“出圈爆红”的可能。地方民生类节目应当在坚持记者操守的基础上，立足本土特色，深耕内容优势，尊重市场趋势，努力在线上平台寻求自身发展的新机遇与新契机。

就如上，[<span id="fn1">覃玉颖（2021.8）</span>](#fn1-black)这段话，总的来说是对《黄金眼》节目的肯定与认同。《黄金眼》的出圈，吸引了一批新生代用户，一定程度上，这类民生新闻节目拓宽新生代用户见闻视域，同时也增强了他们感知社会视阈。从听众视角来看，《黄金眼》反映了不少热点民生话题，是为大众着想、着眼于大众的民生新闻节目。从媒体发展视角，更多是在如今大众生活节奏加快、阅读习惯改变、聚焦分散等问题上，媒体基于当今社会现实的思考与生存发展，这也在宏观上，反映了民生类新闻在后现代社会的嬗变。

## 舆论失焦现象分析，以《1818黄金眼》小张走红一事为例

以[<span id="fn2"> 商锦暄（2021）</span>](#fn2-black)引用[<span id="fn3"> 严利华（2016）</span>](#fn3-black)的定义解释：“由于网络发展，公众知情权、话语权提升，事件中舆论难以被一方主导，使得舆情演变的主体脉络呈现多极化发展以至逐渐偏离事件中心议题的现象。” 简单来说，这里的失焦就是脱离主题。

[<span id="fn2"> 商锦暄（2021）</span>](#fn2-black)分析失焦成因的如下三点是一个结构性的总纲概括，但行文叙述往往是总分结构。而人对事件的总结，通常是分总顺序进行的。这点要注意。

1. “注意力经济”时代与“流量为王”生存法则繁衍舆论失焦的新闻报道。
2. 创新节目内容与病毒式传播渠道是舆论失焦的催化剂。
3. 身处风险社会的受众赋予新闻节目娱乐价值致使新闻导向偏航。

简略提及如上三点的具体分析的关键短语：

1. “注意力经济”、“流量为王”、“颜值”、“快感”、“关注度”、“讨论度”、“货币”、“筹码”
2. “[精英化](https://baike.baidu.com/item/精英文化/11042927?fromtitle=精英化&fromid=16795454&fr=aladdin)到[平民化](https://baike.baidu.com/item/平民化/6949324?fr=ge_ala)“、“严肃性到娱乐性”、“啼笑皆非”、“匪夷所思”、“病毒式传播”、“用户群庞大”、“媒介素养不均”
3. 身处风险社会、对严肃内容的排斥心理、排解忧虑、娱乐方式、获得快感、满足窥私欲

由此，这也可以解释我之前的疑问 “为什么我看同事日常精干处理工作事务，但涉及到电视剧、新闻报道却是人云亦云没多少见解，像个白痴一样？” 总的说来，舆论失焦现象也是必然且常态的现象。用户对新闻标题的猎奇点击，对事件的耐心等待与深入思考已不符合这方面一部分受众用户的期望了。在注意力、精力有限的情况下，这部分受众用户从而转向其他具有话题性、娱乐性的闲谈讨论，爆米花式的休闲娱乐（吃瓜）以此舒缓压力获得快感。

## 新闻网评节目 ——以《睡前消息》为例

从陈家玮（2021）对《睡前消息》76期节目【武汉软件学院清理学生宿舍问题】调查不完善，点评偏颇，引发了许多大学生受众的反对意见和持续争议。可我看了《睡前消息》76期文稿[【为什么奖励主动承认病情的人，会对防疫有好处？】](https://btnews.ktlab.io/btnews/idx/0076/) 中提到的这一事件，引发大多数学生反对，难说。发言权通常往往被占强势的校方所代表，所谓的民族大义、顾及集体利益，往往要分割大多数学生的利益，如今注重自我与个人的时代，这样的做法可未必会得到大多数学生义愤填膺的认同与支持，更多的唤起心中同情而自发提供的帮助。

陈家玮（2021）：《睡前消息》坚持理性、去情绪化思维，挖掘新闻事件深层，平息受众的怒气和怨气，缓解社会矛盾，并从相关事件衍发对基层治理的思考和讨论，取得了不错的效果。这也说明这类网评节目存在的必要性，从一定程度上来说，《睡前消息》的建设性新闻实践构筑了民众与社会、经济学之间开拓性探索的桥梁。立足未来，提出前瞻性解决对策，有促进开启民智的作用。不过，额外强调一点：开启民智目的是让每个人都具备批判性思维，能够自主思考和客观分析问题，不被误导或愚弄；它强调的是社会共同进步的需要，而不是贬低民众的能力或价值。

[<span id="fn4"> 陈家玮（2021）</span>](#fn4-black) 、[<span id="fn5"> 何鑫（2021）</span>](#fn5-black) 两者各自分析了《睡前消息》存在的客观局限，基于《睡前消息》团队规模、新闻多样化的内容，团队专业能力范围存在局限，难以稳控周期更新的节目质量。何鑫对《睡前消息》的态度更近似于学习性的研究，着重于自媒体时评发展的启示方面，较少提出建设性意见。陈家玮则表示：“《睡前消息》作为有一定影响力的自媒体时政评论节目，提倡唯物主义、实事求是，以人民群众为中心的理念，在此次疫情报道中发挥自身优势，与主流媒体错位发展，优势互补，通过解释报道框架，对各类舆论热点事件进行科学理性的分析，引导受众产生积极正面的情绪，缓释社会矛盾，同时提出面向未来的可行性建设性方案，彰显了建设性新闻的问题导向和行动导向，但仍然存在很多不足之处。自媒体工作者需要提升职业素养，加强信源权威，邀请专业智囊把关，改进节目形式，并适当协调好商业营销和建设性新闻的敏感关系，在重大突发性公共事件的建设性报道中更好地发挥自媒体的作用。” 深中肯綮道在自媒体时评的整体积极的发展方向。

## 番外

### ChatGPT 润色：

[<span id="fn4"> 陈家玮（2021）</span>](#fn4-black) 、[<span id="fn5"> 何鑫（2021）</span>](#fn5-black) 分别对《睡前消息》存在的客观局限进行了分析。他们指出，《睡前消息》团队规模有限，导致新闻内容的多样性受到了限制，同时团队在专业能力范围上也存在局限，难以稳定控制节目质量的更新周期。在对《睡前消息》的态度上，何鑫更多地倾向于将其视为学习性研究的对象，主要关注自媒体时评发展的启示方面，较少提出建设性意见。

相比之下，陈家玮则认为，《睡前消息》作为一档有一定影响力的自媒体时政评论节目，秉承了唯物主义、实事求是的理念，并以人民群众为中心。他强调在疫情报道中，《睡前消息》充分发挥了自身优势，与主流媒体形成了错位发展，实现了互补。该节目通过解释报道框架，对各类舆论热点事件进行科学理性的分析，引导受众产生积极正面的情绪，缓和了社会矛盾。同时，节目也提出了面向未来的可行性建设性方案，彰显了建设性新闻的问题导向和行动导向。

然而，陈家玮也意识到，《睡前消息》仍然存在许多不足之处。他认为自媒体工作者需要提升职业素养，加强信源权威，并邀请专业智囊进行把关。此外，改进节目形式以及适当协调商业营销和建设性新闻之间的敏感关系也是必要的措施。尤其在重大突发性公共事件的建设性报道中，自媒体应更好地发挥作用。

综上所述，这段文字描述了陈家玮和何鑫对《睡前消息》存在的客观局限的分析，并呈现出两位作者在态度和观点上的不同。陈家玮强调了节目在疫情报道中的优势和与主流媒体的互补，同时也指出了改进的方向和必要的措施。这段润色后的文字依然表达了自媒体时评的积极发展方向，避免了冗余和歧义的表达。
